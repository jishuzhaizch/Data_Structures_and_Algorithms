## 求和

写一个递归程序求得0到n的和（1到n的和）。

````python
def get_sum(n):
    if n == 0:
        return 0
    return get_sum(n-1) + n
````

````python
n = 10
res = get_sum(n)
res
````

## 阶乘

求解1x2x3x4...xn

````python
def get_factorial(n):
    if n <= 0:
        return False
    if n == 1:
        return 1
    return get_factorial(n-1) * n
````

````python
res = get_factorial(10)
res
````

````python
res = get_factorial(5)
res
````

## 斐波那契数列

求斐波那契数列的第n项

````python
def fibonacci(n):
    if n <= 2:
        return 1
    return fibonacci(n-1) + fibonacci(n-2)
````

````python
res = fibonacci(10)
res
````

## 数学表达式

给定a和b，其中a大于b，有两种操作，分别是乘2和加1操作。求出最小的操作次数的数学表达式。

如：

23 = ((5 * 2 + 1) * 2 + 1)

113 = ((((11 + 1) + 1) + 1) * 2 * 2 * 2 + 1)

解题思路：

首先判断a是否与b相等，相等则退出。

如果a大于2倍的b，则进行乘2处理。

如果a小于2倍的b，则加1。

最终将a做除以2和减1的操作。

````python
def intSeq(a, b):
    if a==b:
        return str(b)
    if a % 2 == 1:  # a为奇数
        return "(" +intSeq(a-1, b)+ " + 1)"
    if a < 2 * b:
        return "(" +intSeq(a-1, b)+ " + 1)"
    return intSeq(a/2 ,b) + " * 2 "
````

````python
a = 113
b = 11
res = intSeq(a, b)
res
````

## 格雷码

格雷码有个特点每次仅仅变换一位：

如：

0000 -> 0001 -> 0011 -> 0010
   
  enter1  enter2  exit1 ...


要求打印格雷码的操作

````python
def move_idx(n):
    if n == 0:
        return 
    move_idx(n-1)
    print(n)
    move_idx(n-1)
````

````python
move_idx(5)
````

解题思路：

1. 利用一个标志位进行状态的选择
2. 当该数值第一次出现的时候，为enter，再一次出现的时候为exit

````python
def grey_code_move(n ,flag):
    if n == 0:
        return 
    grey_code_move(n-1, True)
    print(n, "enter") if flag else print(n, "exit")
    grey_code_move(n-1, False)
````

````python
grey_code_move(3,True)
````

## 汉诺塔

时间复杂度O（2^n）

n个铁饼，需要移动2^n-1个

````python
def hanoi(n, L, M, R):
    if n == 1:
        print("From " + L + " To " + R)
    else:
        hanoi(n-1, L, R, M) # 处理前n-1层
        hanoi(1, L, M, R) # 处理最后一层，就是n为1的情况
        hanoi(n-1, M, L, R)
````

````python
n = 3
hanoi(n, "left", "middle", "right")
````

## 子集问题

给定一个列表，其中没有重复的数字。生成所有元素的所有子集的结果。

例如：
\[1,2,3\]

$[[],[1],[2],[3],[1,2],[1,3],[2,3],[1,2,3]]$

解决方法：
1. 利用Python自带的copy\[:\]方法进行处理。
如何处理？
先初始化一个$[[]]$,在遍历原先的list,从中抽取一个元素，添加到result的元素后面，形成新的子集集合中的元素，再将其添加进来。

也就是：
```
[[]] --> [[], [1]] --> [[],[1],[2],[1,2]] --> [[], [1], [2], [1,2],[3], [1,3], [2,3], [1,2,3]]
```

````python
 def get_subset(arr):
    result_list = [[]]
    for num in arr: # 遍历列表中的每一个元素
        for subset in result_list[:]:  # copy一个result_list的副本
            new_subset = subset + [num]
            result_list.append(new_subset)
    return result_list
````

````python
arr = [1,2,3]
res = get_subset(arr)
res
````

解法2：

利用回溯法进行子集的遍历与生成子集集合，首先遍历第一个节点树，可以进行选择或不选择后续节点的操作，知道将后续节点情况均遍历完成后，可以对剩余的节点数进行下一步的操作



````python
def backtracking_subset(arr):
    inter_result = []
    result_list = []
    helper(result_list, inter_result, arr, 0)
    return result_list

def helper(result_list, inter_result, arr, position):
    result_list.append(inter_result[:]) # 添加inter_result的copy副本
    for idx in range(position, len(arr)):
        inter_result.append(arr[idx])
        helper(result_list, inter_result, arr, idx + 1)
        inter_result.pop()
````

````python
res = backtracking_subset(arr)
res
````

````python
arr2 = [1,2,2]
res = backtracking_subset(arr2)
res
````

## 子集Ⅱ

如果子集中有重复的元素就会出现结果中出现大量的重复。如何避免这类问题的产生？

解法：

分为两步走

step1：第一步对列表中的元素进行排序。

step2：针对当前处理的元素和上一个处理的元素一致的情况可以进行剪枝操作。

````python
def get_subset2(arr):
    result_list = [[]]
    for idx, num in enumerate(arr, -1):
        for subset in result_list[:]:
            new_subset = subset + [num]
            result_list.append(new_subset) if new_subset not in result_list else None
    return result_list
````

````python
res = get_subset2(arr2)
res
````

````python
def backtracking_subset2(arr):
    arr.sort()
    inter_result = []
    result_list = []
    helper2(result_list, inter_result, arr, 0)
    return result_list

def helper2(result_list, inter_result, arr, position):
    result_list.append(inter_result[:])
    for idx in range(position, len(arr)):
        if (idx != position and arr[idx - 1] == arr[idx]):
            continue
        inter_result.append(arr[idx])
        helper2(result_list, inter_result, arr, idx + 1)
        inter_result.pop()
````

````python
arr2 = [1,2,2]

res = backtracking_subset2(arr2)
res
````

## 排列组合

排列组合问题

给定一个数组：abc

给出这个数组的全排列：

abc,acb,bac,bca,cab,cba

解法：

采用递归的方式，在每层顺序的遍历该层的元素，将所有层的结果放在一起，组成最后的结果

````python
def perm(result, arr):
    if len(arr) == 0:
        print(result)
        
    for i in range(len(arr)):
        perm(result + str(arr[i]), arr[:i] + arr[i+1:])
````

````python
nums = [1, 2, 3]
perm("", nums)
````

唯一性问题，如果列表中出现了重复的元素。如何进行全排列的输出

````python
nums = [1,2,2]
perm("", nums)
````

````python
def perm2(result, arr):
    arr.sort()
    if len(arr) == 0:
        print(result)
    for i in range(len(arr)):
        if (i != 0 and arr[i-1] == arr[i]):
            continue
        perm2(result + str(arr[i]), arr[:i] + arr[i+1:])
````

````python
nums = [1,2,2]
perm2("", nums)
````

从n个元素中抽取k个作为最后的结果，即$C_n^k$的问题

````python
def permk(result, arr, k):
    if k == 0:
        print(result)
    for i in range(len(arr)):
        permk(result + str(arr[i]), arr[:i] + arr[i+1:], k-1)
````

````python
nums = [1,2,3,4]
permk("",nums, 2)
````

## combination sum

从list中抽取k个值和为target

````python
def comb(arr, t):
    result = []
    tmp = []
    combhelper(result, tmp, arr, t, 0)
    return result

def combhelper(result, tmp, arr, remains, start):
    if remains < 0:
        return 
    if remains == 0:
        result.append(tmp[:])
        
    else:
        for i in range(start, len(arr)):
            tmp.append(arr[i])
            combhelper(result, tmp, arr, remains - arr[i], i + 1)
            tmp.pop()
````

````python
candidates = [2,3,6,7]
t = 7
comb(candidates, t)
````
